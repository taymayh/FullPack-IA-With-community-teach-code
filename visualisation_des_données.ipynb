```python
lien = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
```

```python
titanic = pd.read_csv(lien)
```

* Survived : Indique si le passager a survcu (1) ou non (0)
* Pclass : Classe du billet (1, 2, ou 3)
* Name : Nom du passager
* Sex : Sexe du passager
* Age : ge du passager
* SibSp : Nombre de frres/surs et conjoints  bord
* Parch : Nombre de parents/enfants  bord
* Ticket : Numro du billet
* Fare : Tarif pay pour le billet
* Cabin : Numro de la cabine
* Embarked : Port d'embarquement (C = Cherbourg; Q = Queenstown; S = Southampton)




```python
df = titanic
```

```python
df
```

```python
# Filtrer les survivants (Survived == 1) et les non-survivants (Survived == 0)
survived_data = df[df['Survived'] == 1]
not_survived_data = df[df['Survived'] == 0]

# Compter le nombre d'hommes et de femmes parmi les survivants
survived_gender_count = survived_data['Sex'].value_counts()

# Compter le nombre d'hommes et de femmes parmi les non-survivants
not_survived_gender_count = not_survived_data['Sex'].value_counts()

# Afficher les rsultats
print("\nNombre de survivants par sexe :")
print(survived_gender_count)

print("\nNombre de non-survivants par sexe :")
print(not_survived_gender_count)
```

```python

# Crer une figure et des sous-graphiques
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Graphique des survivants par sexe
sns.barplot(x=survived_gender_count.index, y=survived_gender_count.values, ax=ax[0], palette='viridis')
ax[0].set_title("Survivants par sexe")
ax[0].set_ylabel("Nombre de survivants")
ax[0].set_xlabel("Sexe")

# Graphique des non-survivants par sexe
sns.barplot(x=not_survived_gender_count.index, y=not_survived_gender_count.values, ax=ax[1], palette='viridis')
ax[1].set_title("Non-survivants par sexe")
ax[1].set_ylabel("Nombre de non-survivants")
ax[1].set_xlabel("Sexe")

# Afficher les graphiques
plt.tight_layout()
plt.show()
```
